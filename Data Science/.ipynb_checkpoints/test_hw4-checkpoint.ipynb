```python
from lxml import html
import requests
from datetime import date
```

```python
import unicodedata
```

```python
def lenta_news(number_of_news):
    
    news_by_lenta = []
    
    header = {
    'USER-AGENT': \
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
    
    url = "https://lenta.ru/"
    
    response = requests.get(url, headers=header)
    
    root = html.fromstring(response.text)
    
    number_of_news = number_of_news + 1 
    
    for i in range(2, number_of_news):
        lenta_name_draft = root.xpath(f'//*[@id="root"]/section[2]/div/div/div[2]/div[1]/section/div/div[{i}]/a/text()')
        lenta_date = root.xpath(f'//*[@id="root"]/section[2]/div/div/div[2]/div[1]/section/div/div[{i}]/a/@href')
        lenta_href = root.xpath(f'//*[@id="root"]/section[2]/div/div/div[2]/div[1]/section/div/div[{i}]/a/@href')
        
        lenta_name = []
    
        for el in lenta_name_draft:
            el = unicodedata.normalize("NFKD", el)
            lenta_name.append(el)
            
        lenta_href = str(lenta_href)
        lenta_name = str(lenta_name)
        elements = ['[', ']', "'"]
        
        if "lenta" not in lenta_href:
            for element in elements:
                lenta_href = lenta_href.replace(element, '')
            lenta_href = 'https://lenta.ru' + lenta_href
        
        for element in elements:
            lenta_name = lenta_name.replace(element, '')
        
        lenta_date = str(lenta_date)
        index = lenta_date.find("2021")
        lenta_date = lenta_date[index: index+10]
        lenta_date = lenta_date.replace(u'/', u'-')
        
        news_by_lenta.append({
            'Name': lenta_name,
            'Href': lenta_href,
            'Date': lenta_date,
            'Source': 'Lenta.ru'
        })
        
    return news_by_lenta    
```

```python
lenta_news(7)
```

```python
def yandex_news():
    
    news_by_yandex = []
    
    header = {
    'USER-AGENT': \
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
    
    url = "https://yandex.ru/news"
    
    response = requests.get(url, headers=header)
    
    root = html.fromstring(response.text)
    
    xpath_yandex_title = '//div[@class="mg-grid__row mg-grid__row_gap_8 news-top-flexible-stories news-app__top"]\
/div/article/div/div/a/h2/text() | //*[@id="neo-page"]/div/div[2]/div/div[1]/div[1]/div[1]/article/div[2]/a/h2/text()'
    
    xpath_yandex_href = '//div[@class="mg-grid__row mg-grid__row_gap_8 news-top-flexible-stories news-app__top"]\
/div/article/div/div/a/@href | //*[@id="neo-page"]/div/div[2]/div/div[1]/div[1]/div[1]/article/div[2]/a/@href'
    
    yandex_title_draft = root.xpath(xpath_yandex_title)
    yandex_href = root.xpath(xpath_yandex_href)
    
    yandex_title = []
    
    for el in yandex_title_draft:
            el = unicodedata.normalize("NFKD", el)
            yandex_title.append(el)
    
    data = date.today()
    
    for i in range(len(yandex_title)):
        news_by_yandex.append({
            'Name': yandex_title[i],
            'Href':  yandex_href[i],
            'Date': date,
            'Source': 'Yandex.ru'
        })
    
    return news_by_yandex
```

```python
def news_mail_ru():
    
    news = []    
    
    headers = {'USER-AGENT': \
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
    
    keys = ('title', 'date', 'link')
    date_format = '%Y-%m-%dT%H:%M:%S%z'
    

    link_mail_ru = 'https://mail.ru/'
    
    request = requests.get(link_mail_ru, headers=headers)
    root = html.fromstring(request.text)
    
    news_links = root.xpath('''(//div[@class =  "news-item o-media news-item_media news-item_main"]  |  
                                //div[@class =  "news-item__inner"])
                                /a[contains(@href, "news.mail.ru")]/@href''')
    
    news_text = root.xpath('''(//div[@class =  "news-item o-media news-item_media news-item_main"]//h3  |  
                               //div[@class =  "news-item__inner"]/a[contains(@href, "news.mail.ru")])
                               /text()''')
    
    for i in range(len(news_text)):
        news_text[i] = news_text[i].replace(u'\xa0', u' ')
    
    news_links_temp = []
    for item in news_links:
        item = item.split('/')
        news_links_temp.append('/'.join(item[0:5]))
        
    news_links = news_links_temp
    del(news_links_temp)
    
    news_date = []
    
    for item in news_links:
        request = requests.get(item, headers=headers)
        root = html.fromstring(request.text)
        date = root.xpath('//span[@class="note__text breadcrumbs__text js-ago"]/@datetime')
        news_date.extend(date)
        
    for i in range(len(news_date)):
        news_date[i] = datetime.strptime(news_date[i], date_format)
        
    for item in list(zip(news_text, news_date, news_links)):
        news_dict = {}
        for key, value in zip(keys, item):
            news_dict[key] = value
        
        news_dict['source'] = 'mail.ru'
        news.append(news_dict)
    
    return news
```
